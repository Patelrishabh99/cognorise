# BREAST CANCER CLASSIFICATION

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import sklearn.datasets
```

```python
X=data_frame.drop(columns='label',axis=1)
Y=data_frame['label']
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train.shape
```

```python
Y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

ACCURACY IN TRAIN DATA

```python
model.score(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
```

**classification** **report**

```python
print("the classification report is\n\n",classification_report(Y_test,y_pred))
```

**confusion** **matrix**

```python
print("the confusion matrix is:\n\n",confusion_matrix(Y_test,y_pred))
```

**accuracy** **on** **test** **data**

```python
print("the accuracy score is::\n\n",accuracy_score(Y_test,y_pred)*100)
```

**THE** **ACCURACY** **SORE** **ON** **TEST** **DATA** IS **96.4**
